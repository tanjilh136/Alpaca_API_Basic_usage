```python
#assign it into a input field
symbol=input("Enter the ticker symbol of Desired Company\n").upper()

#assign this function into the cimpany name button 
def find_companyname(symbol):
    global aapl_asset
    aapl_asset=api.get_asset(symbol)
    return aapl_asset.name

find_companyname(symbol)
```

```python
#call this function to the shortable field 
def find_shortable(symbol):
    global aapl_asset
    aapl_asset=api.get_asset(symbol)
    if aapl_asset.shortable:
        return 'S'
    else:
        return 'N'
    

find_shortable(symbol)
```

```python
#This will start when the program intiates, will not go in any function ..... it will refresh when program will be closed
data_loader=[]

def on_open(ws):
    print("--------------Starting The Data Collecting process----------------")
    auth_data = {
        "action": "auth",
        "params": "Zay2cQZwZfUTozLiLmyprY4Sr3uK27Vp"
    }

    ws.send(json.dumps(auth_data))

    channel_data = {
        "action": "subscribe",
        "params": f"T.{symbol}"
    }

    ws.send(json.dumps(channel_data))
    


def on_message(ws, message):
    message=json.loads(message)
    print("received a message")
    print(message)
    data_loader.append(message)
    

def on_close(ws):
    print("closed connection & Stored")

socket = 'wss://socket.polygon.io/stocks'

ws = websocket.WebSocketApp(socket, on_open=on_open, on_message=on_message, on_close=on_close)
ws.run_forever()


#it will save the data to a csv 
def data_save(data_loader):
    time=[]
    price=[]
    for i in range(3,len(data_loader)):
        time.append(data_loader[i][0]['t'])
        price.append(data_loader[i][0]['p'])
    df=pd.DataFrame()
    df['UNIX TIME-STAMP']=time
    df['Price']=price
    df.to_csv('price_with_time.csv',index=False)
    return "stored Into a csv"
    
data_save(data_loader)
```

```python

#This will return  the quantity 
df_=pd.read_csv('price_with_time.csv')
def formula1(df_):
    now=datetime.now().timestamp()
    end=(datetime.now()+timedelta(seconds=(60*10))).timestamp()
    ds=df_.loc[(df_['UNIX TIME-STAMP'] > now) & (df_['UNIX TIME-STAMP'] <= end), ['UNIX TIME-STAMP', 'Price']]
    
    now_2=end
    end2=end+600
    ds1=df_.loc[(df_['UNIX TIME-STAMP'] > now_2) & (df_['UNIX TIME-STAMP'] <= end2), ['UNIX TIME-STAMP', 'Price']]
    
    now_3=end2
    end3=end2+600
    ds2=df_.loc[(df_['UNIX TIME-STAMP'] > now_3) & (df_['UNIX TIME-STAMP'] <= end3), ['UNIX TIME-STAMP', 'Price']]
    
    A=sum(ds['Price'])/len(ds['Price'])
    B=sum(ds1['Price'])/len(ds1['Price'])
    C=sum(ds2['Price'])/len(ds2['Price'])
    
    return ((A*3)+(B*2)+C)/120
quantity=formula1(df_)
```

```python
#submit_order_for_buy
order_type=input("Enter the market order").lower()

def buy_order(order_type,symbol,quantity):
    
    if order_type=='market':
        api.submit_order(
            symbol=symbol,
            qty=quantity,
            side='buy',
            type=order_type,
            time_in_force='gtc'
        )
    elif order_type=='limit':
        limiting_price=float(input("Enter the limiting price"))
        api.submit_order(
            symbol=symbol,
            qty=quantity,
            side='buy',
            type=order_type,
            time_in_force='opg',
            limit_price=limiting_price
        )

buy_order(order_type,symbol,quantity)
    
```

```python
#submit_order_for_SELL
order_type=input("Enter the market order").lower()

def sell_order(order_type,symbol,quantity):
    if order_type=='market':
        api.submit_order(
            symbol=symbol,
            qty=quantity,
            side='sell',
            type=order_type,
            time_in_force='gtc'
        )
    elif order_type=='limit':
        limiting_price=float(input("Enter the limiting price"))
        api.submit_order(
            symbol=symbol,
            qty=quantity,
            side='sell',
            type=order_type,
            time_in_force='opg',
            limit_price=limiting_price
        )
sell_order(order_type,symbol,quantity)
```
